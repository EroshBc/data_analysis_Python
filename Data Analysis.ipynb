```python
#select only races in USA 50km or 50 miles
```

```python
#Select only the USA
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi','50miles'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') ]
df2
```

```python
#remove USA name 
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean athlete age
```

```python
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2.head()
```

```python
#drop columns: Athlete club, Athlete country, athlete year of birth, athlete age category
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country','Athlete year of birth','Athlete age category'],axis = 1)
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
#check for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
# reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fix data types
```

```python
df2.dtypes
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#rename columns
```

```python
df2 = df2.rename(columns = {
    'Year of event' : 'year',
    'Event dates' : 'race_day',
    'Event name' : 'race_name',
    'Event distance/length' : 'race_length' ,
    'Event number of finishers' : 'race_finishers_num',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender' ,
    'Athlete average speed' : 'athelete_average_speed' ,
    'Athlete ID' : 'athelete_ID',
    'Athlete_age' : 'age' 
    
})
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 =df2[['race_day','race_name','race_length','race_finishers_num','athelete_ID','athlete_gender','athlete_age','athelete_average_speed','athelete_ID','race_year']]
```

```python
#which age groups are best in the 50m race 
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athelete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count > 19').head(15)
```

```python
#Seasons for the data -> slower in summer than winter
#spring 3-5
#summer 6-8
#fall 9 - 11
#winter 12 - 2
#split between to  decimals
```

```python
df3.dtypes
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athelete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
